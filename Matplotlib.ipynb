```python
import pandas as pd
import matplotlib.pyplot as pt

wdata = pd.DataFrame({"Year":[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024],
                      "High_Temp(Degree)":[28,27,29,28,27,28,29,30,28,30,28,27,30,30,29,31],
                      "Low_Temp(Degree)":[21,22,23,21,21,22,23,23,22,23,22,20,22,16,18,18],
                     })
wdata
```

```python
x = wdata["Year"]
y = wdata["High_Temp(Degree)"]
y1 =wdata["Low_Temp(Degree)"]
pt.figure(figsize = (12,4))
pt.plot(x,y,label = "High Temp (Degree)",color = "#E6B200")
pt.plot(x,y1,label = "Low Temp (Degree)",color = "#82BEE8")
pt.title("Udaipur November Weather Year By Year",fontsize = 15)
pt.xlabel("Years",fontsize = 11)
pt.ylabel("Temperature In Degree",fontsize = 11)
pt.legend()
pt.xticks(x)
pt.annotate("lowest",xy = (2022,16),xytext = (2023,20),arrowprops = {"facecolor":"k","width":0.2,"mutation_scale":1},
             bbox={"facecolor":"#D4CDF7"})
pt.annotate("Highest",xy = (2024,31),xytext = (2023,26),arrowprops = {"facecolor":"k","width":0.2,"mutation_scale":1},
             bbox={"facecolor":"#D4CDF7"})
pt.grid(alpha = 0.3)
pt.savefig("Weather_November.png",dpi= 2500)
pt.show()
```
